```python
#Siya Behl
#The program I have created below combines two text files which were uploaded/mounted in the above blocks of code
#These two files are then combined to create a master list of names which a user can search through
#The code functions similar to Google in that it allows a user to input a series of conditions for an inquiry and returns an output accordingly

#Function compiles names from a text file to create a master list
def create_master_list(file_path):
    names = []
    with open(file_path, 'r') as file:
        for line in file:
            names.append(line.strip())
    return names

#Function searches for names in the master list
def search_names(master_list, gender, exact_match, starts_with):
    results = []
    for name in master_list:
        #Checks if the name starts with the given character
        if name.lower().startswith(starts_with.lower()):
            if gender.lower() == 'all' or (gender.lower() == 'boy' and name in boy_names) or (gender.lower() == 'girl' and name in girl_names):
                if exact_match.lower() == 'yes' and name.lower() == starts_with.lower():
                    #If an exact match is requested, the name is only added if it is an exact match
                    results.append(name)
                elif exact_match.lower() == 'no':
                    #If an exact match is not requested, the name is added to the results of the search
                    results.append(name)
    return results


#Function displays the search results
def display_results(results):
    if len(results) == 0:
        #If there are no results for the search, the following error message is given
        print("No matching names found.")
    else:
        #Else, the names that were found that match the search requests will be printed
        print("Matching names:")
        for name in results:
            print(name)

#Function that initiates the main portion of the name search program
def main():
    #File path for BoyNames.txt
    boy_names_file_path = "/content/BoyNames.txt"
    #File path for GirlNames.txt
    girl_names_file_path = "/content/GirlNames.txt"
    #Initiates a list that will hold both the girl and boy names
    master_list = []
    #Create the master list by combining names from both files
    boy_names = create_master_list(boy_names_file_path)
    girl_names = create_master_list(girl_names_file_path)

    #While loop to print the main menu until a valid choice is provided by the user
    while True:
        print("\nMenu:")
        #ADDITIONAL ENHANCEMENT: Allows the user to initiate the creation of the master list
        print("1. Create Master List")
        print("2. Search for Names")
        print("3. Exit")
        #Prompts for user input on desired menu option
        choice = input("Enter your choice (1-3): ")

        #Choice 1 initiates the creation of the master list which combines text from both BoyNames and GirlNames files
        if choice == '1':
            #Create the master list by combining names from both files
            master_list = boy_names + girl_names
            print("Master list created.")

        #Choice 2 initiates ONLY AFTER the master list has been initiated for creation by the user and allows for the search to begin
        elif choice == '2':
            #ADDITIONAL ENHANCEMENT: Ensures that the user has created the master list before proceeding to search
            if not master_list:
                print("Please create the master list first.")
            #If the master list has been created by the time the user selects 2, then the following questions are asked
            else:
                gender = input("Enter the gender (Boy, Girl, or All): ")
                exact_match = input("Do you want an exact match? (Yes/No): ")
                starts_with = input("Enter the name to search starting with: ")
                if gender.lower() == 'boy':
                    results = search_names(boy_names, gender, exact_match, starts_with)
                elif gender.lower() == 'girl':
                    results = search_names(girl_names, gender, exact_match, starts_with)
                else:
                    results = search_names(master_list, gender, exact_match, starts_with)
                display_results(results)

        #Choice 3 allows the user to exit the program if they do not wish to commence a search
        elif choice == '3':
            print("Exiting the program.")
            break

        #If none of the above choices are selected, or an invalid choice is input, then the following message is printed
        else:
            print("Invalid choice. Please try again.")

#Calls upon the main function to compile the code
main()
```
